```python
import pandas as pd
import numpy as np
import os
from config import *
```

```python
csvfile = "./cshares.csv"
df = pd.read_csv(csvfile,dtype=object)
df.head()
```

```python
df.price = df.price.astype(float)
df.price2 = df.price2.astype(float)
df['cprice']=df.price2-df.price
```

```python
df.head()
```

```python
shares = 100
df['cost'] = 100 * df.price
df.head()
```

```python
df['gains'] = shares * (df.cprice)
df.head()
```

```python
df.groupby('file_date').agg({'cost':'sum', 'gains': 'sum'}).reset_index().rename(columns={'cost':'Total Cost', 'gains':'Total Gains'})
```
