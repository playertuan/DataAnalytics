```python
import requests
import xml.etree.ElementTree as ET
import pandas as pd
```

```python
def xml2df(xml_data):
    root = xml_data
    all_records = []
    for i, child in enumerate(root):
        record = {}
        for subchild in child:
            if subchild.tag == "cusip":
                record["cusip"] = subchild.text
            if subchild.tag == "nameOfIssuer":
                record["name"] = subchild.text
            if subchild.tag == "value":
                record["marketvalue"] = float(subchild.text)
            if subchild.tag == "shrsOrPrnAmt":
                for grandchild in subchild:
                    if grandchild.tag == "sshPrnamt":
                        record["shares"] = float(grandchild.text)
            #record[subchild.tag] = subchild.text
        all_records.append(record)
    return pd.DataFrame(all_records)
```

```python
response = requests.get('https://www.sec.gov/Archives/edgar/data/1067983/000095012317007953/form13fInfoTable.xml')
xml_response = str(response.content.decode('utf-8'))
test =xml_response.replace(' xmlns="http://www.sec.gov/edgar/document/thirteenf/informationtable"','')
test2 = test.replace(' xmlns:xsi="http://www.w3.org/2001/XMLSchema-instance"','')
#print(test2)
#print(response.content)
root = ET.fromstring(test2)
#print(root[0])
#print(root[1])
df = xml2df(root)
df.head(100)
```
